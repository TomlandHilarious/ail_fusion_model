```python
# Filter for VideoStim1_started and VideoStim1_ended markers
start_marker = eeg_df[eeg_df['MarkerName'] == 'VideoStim1_started'].index
end_marker = eeg_df[eeg_df['MarkerName'] == 'VideoStim1_stopped'].index

if len(start_marker) > 0 and len(end_marker) > 0:
    # Calculate length (number of samples)
    length = end_marker[0] - start_marker[0]
    print(f"Length (number of samples): {length}")
    
    # Divide by 256 (sampling rate) to get time in seconds
    time_seconds = length / 256
    print(f"Time in seconds: {time_seconds:.2f}")
else:
    print("Could not find VideoStim1_started and/or VideoStim1_stopped markers")
```
